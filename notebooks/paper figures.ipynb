```python
from collections import Counter
counter = Counter()
no_tags = 0
with_tags = 0
for t, c, s, fp, labels in train_dataset.index:
    if s == 0:
        if sum(labels) == 0:
            no_tags += 1
        else:
            with_tags += 1
            for idx, l in enumerate(labels):
                if l != 0:
                    counter[idx] += 1
```
